### RETAIL ANALYSIS

### The Analysis and Visualizations below is based on the Top ten Smallest prices for Retail Market Type. N.B- No comparative analysis and visualization can be made for the largest prices for retail market

```python
ret_mrt = dz[dz["market_type"] == "Retail"]
ret_mrt
```

```python
ret_mrt[["country","state","market","produce","quantity","month","year","price"]]
```

```python
# least Price
Tr = ret_mrt[["country","state","market","produce","quantity","month","year","price"]].nsmallest(n=10, columns= "price")
Tr
```

```python
plt.figure(figsize=(10,6))
plt.pie(Tr.price, autopct="%0.1f%%", normalize=True, labels = Tr.country) 
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.pie(Tr.month, autopct="%0.1f%%", normalize=True, labels = Tr.market) 
plt.show()
```

```python
x = Tr.market
y = Tr.price
colors = ("red","blue", "green", "yellow", "pink","black")

plt.figure(figsize=(12,6))
plt.bar(x,y, color = colors)
plt.title("Market Prices")
plt.xlabel("Market")
plt.ylabel("Prices")
plt.xticks(rotation=90)
plt.show()
```

```python
x = Tr.produce
y = Tr.price
colors = ("red","blue", "green", "yellow", "pink","black")

plt.figure(figsize=(12,6))
plt.bar(x,y, color = colors)
plt.title("Product Prices")
plt.xlabel("Produce")
plt.ylabel("Prices")
plt.xticks(rotation=90)
plt.show()
```

```python
x = Tr.market
y = Tr.month
colors = ("red","blue", "green", "yellow", "pink","black")

plt.figure(figsize=(12,6))
plt.bar(x,y, color = colors)
plt.title("Count of month per market")
plt.xlabel("Market")
plt.ylabel("Month")
plt.xticks(rotation=90)
plt.show()
```

# Summary of Key Insights

Exploratory Data Analysis was carried out on the dataset using the describe function
to determine the mean values, count, standard deviation, minimum, maximum, (25th, 50th and 75th)percentile
for the numerical columns in the datasets which include the month, year and the quantity. 
Three peculiar market types were discovered using the Len function and these 
market types (Wholesale, Producer and Retail) were segmented for analysis. 
For the retail markets, analysis and charts were drawn for the top ten smallest 
prices in the retail market segment which captured the following countries; Gambia, Libya,
Zambia and Zimbabwe in this category. Pie Charts were made for the countries with the lowest
prices as well as market area for each month within those smallest prices. 
Similarly, bar chart were made to establish the relationship between market and prices,
produce and prices as well as market and months. Analysis was also made on the producer 
market segment where the top ten largest prices were obtained. The top ten largest prices
in the producer segment revealed only Madagascar as a country was in that category, bar chart
of different states in Madagascar was shown against the prices as well as states and month relationship. 
Finally, analysis was made for the top ten largest price in the Wholesale market segment in which 
Cameroon was the only African country captured in this market segment. 
Bar charts were used to show the relationships between the market and price as well as market and month respectively.


Furthermore, Zimbabwe as an African country was extracted from the entire datasets using the loc function
and visualizations were made on it. The loc function was used to determine the starting index and last 
index numbers of the extracted dataset Zimbabwe used for Visualizations. Bar charts, pie charts, scatter
plot, line charts and histogram were used to show various relationships between states, markets, produce, 
months and prices in Zimbabwe as an African country. For the entire dataset, count plots were made for countries
and year using bar charts as well as market type and month using seaborn respectively.
